```python
    def createEmptyFile(path):
        with open(path, 'a') as file:
            os.utime(path, None)
        return file
```

```python
def process_txt_to_csv(my_file_to_process, my_new_file, my_directory):
    print('Processing {}'.format(my_file_to_process))
    file_csv = my_directory+'/'+my_new_file+'.csv'
    if os.path.exists(file_csv):
        os.remove(file_csv)
    f_csv = createEmptyFile(file_csv)
    head = 'file,method,complexity' + '\n'
    f_csv = open(file_csv,'w')
    f_csv.write(head)
    f_csv.close()
    with open(my_file_to_process, 'r', encoding='utf-8') as f_csv: 
        for line in f_csv:
            my_content = line.split('::')
            if (len(my_content) == 2):
                parte1 = my_content[0].split()
                if (len(parte1) > 1):
                    my_file = parte1[0]
                    my_file = my_file.split(':')[0]
                parte2 = my_content[1].split(',') 
            if (len(parte2) > 2):
                my_method = parte2[0].split()[0]
                if (len(parte2[1]) > 0):
                    my_method_ccn = parte2[1]
                    my_method_ccn = my_method_ccn.split()
                    my_method_ccn = my_method_ccn[0]
            my_tuple = (my_file, my_method, my_method_ccn)
            with open(file_csv, 'a', encoding='utf-8') as f_csv:
                my_line = my_tuple[0] + ',' + my_tuple[1] + ',' + my_tuple[2] + '\n'
                f_csv.write(my_line)
            print(my_tuple)
            print('File {}.csv updated in {} with success!'.format(my_new_file, my_directory))
```
