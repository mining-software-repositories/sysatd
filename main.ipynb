```python
from pydriller import RepositoryMining
import statistics
```

```python
def dictionaryWithAllCommmits(repository):
    dictionaryAux = {}
    for commit in RepositoryMining(repository).traverse_commits():
        commitAuthorNameFormatted = '{}'.format(commit.author.name)
        commitAuthorDateFormatted = '{}'.format(commit.author_date)
        listFilesModifiedInCommit = []
        for modification in commit.modifications:
            itemMofied = '{}'.format(modification.filename)
            listFilesModifiedInCommit.append(itemMofied)
        dictionaryAux[commit.hash] = [commitAuthorNameFormatted, commitAuthorDateFormatted, listFilesModifiedInCommit] 
    return dictionaryAux
```

```python
def myCommits(repository):
    my_dict = {}
    print('Analisando o repositório ' + repository + '...')
    for commit in RepositoryMining(repository).traverse_commits():
        listFilesModifiedInCommit = []
        for modification in commit.modifications:
            itemMofied = '{}'.format(modification.filename)
            listFilesModifiedInCommit.append(itemMofied)
        my_dict[commit.hash] = listFilesModifiedInCommit
    print('concluído')
    return my_dict
```

```python
repository = '/Users/armandosoaressousa/git/hadoop'
v1 = 'hadoop-branch-2.0.5'
v2 = 'hadoop-branch-3.0.0'

hadoopv1 = repository + '/' + v1
hadoopv2 = repository + '/' + v2

my_dict_hadoopv1 = myCommits(hadoopv1)
my_dict_hadoopv2 = myCommits(hadoopv2)
```
